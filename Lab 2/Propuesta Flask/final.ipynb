# Proyecto final

El objetivo de este proyecto es desarrollar un sistema de gestión de tareas en Python que permita a los usuarios organizar y administrar sus actividades diarias de manera eficiente. El sistema deberá incorporar conceptos y técnicas aprendidos en varias lecciones, brindando una oportunidad para aplicar de manera práctica los conocimientos adquiridos.

### Funcionalidades Requeridas:

1. Agregar Tareas:

* Los usuarios deberán poder agregar nuevas tareas especificando un título, una descripción y la fecha de vencimiento.

2. Visualizar Tareas Pendientes:

* Proporcionar una opción para que los usuarios vean una lista de sus tareas pendientes ordenadas por fecha de vencimiento.

3. Modificar Tareas:

* Permitir a los usuarios realizar modificaciones en las tareas existentes, como cambiar la descripción o la fecha de vencimiento.

4. Marcar Tareas como Completadas:

* Implementar una función que permita a los usuarios marcar una tarea como completada. La tarea deberá conservar un historial de su estado.

5. Guardar y Cargar Tareas:

* Utilizar archivos para guardar la lista de tareas de manera persistente entre ejecuciones del programa. Implementar funciones de lectura y escritura de archivos.

6. Exportar e Importar Tareas en Formato JSON:

* Brindar la posibilidad de exportar todas las tareas a un archivo JSON.
* Permitir a los usuarios importar tareas desde un archivo JSON para cargar o respaldar su lista.

7. Filtrar Tareas:

* Implementar funciones que permitan a los usuarios filtrar tareas según diferentes criterios, como fecha de vencimiento o estado.

8. Visualizar Tareas por Fecha Actual:

* Mostrar las tareas pendientes para el día actual al iniciar el programa.

9. Estadísticas (Opcional):

* Proporcionar estadísticas sobre la productividad del usuario, como la cantidad promedio de tareas completadas por día.

10. Interfaz Gráfica de Usuario (Opcional):

* Si decides implementar una interfaz gráfica de usuario, esta deberá facilitar la interacción del usuario con las funciones mencionadas.

11. Recordatorios y Alertas (Opcional):

* Permitir a los usuarios establecer recordatorios para ciertas tareas y mostrar alertas cuando sea necesario.

* Código fuente del programa.
* Archivo de texto con instrucciones de ejecución.
* Informe breve destacando decisiones de diseño, desafíos enfrentados y posibles mejoras futuras.
